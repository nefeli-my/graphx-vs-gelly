```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy  as np
```

```python
# Plotting for PageRank execution on the 500K Dataset
# Adjust accordingly for more datasets

# Running Times

barabasi_graphx_2g = len(pd.read_csv('PageRank_500K_barabasi_2g_WR.csv').index)
barabasi_graphx_4g = len(pd.read_csv('PageRank_500K_barabasi_4g_WR.csv').index)

erdos_graphx_2g = len(pd.read_csv('PageRank_500K_erdos_2g_WR.csv').index)
erdos_graphx_4g = len(pd.read_csv('PageRank_500K_erdos_4g_WR.csv').index)

barabasi_flink_2g = len(pd.read_csv('PageRank_500K_barabasi_Flink_2g_WR.csv').index)
barabasi_flink_4g = len(pd.read_csv('PageRank_500K_barabasi_Flink_4g_WR.csv').index)

erdos_flink_2g = len(pd.read_csv('PageRank_500K_erdos_Flink_2g_WR.csv').index)
erdos_flink_4g = len(pd.read_csv('PageRank_500K_erdos_Flink_4g_WR.csv').index)
```

```python
labels = ['Gelly-Erdos', 'GraphX-Erdos', 'Gelly-Barabasi', 'GraphX-Barabasi']
g2 = [erdos_flink_2g, erdos_graphx_2g, barabasi_flink_2g, barabasi_graphx_2g]
g4 = [erdos_flink_4g, erdos_graphx_4g, barabasi_flink_4g, barabasi_graphx_4g]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, g2, width, label='2GB', color='orange')
rects2 = ax.bar(x + width/2, g4, width, label='4GB', color='lightblue')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Running Time (s)')
ax.set_title('PageRank - 500K nodes')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
# CPU Usage line plotting 

graphx_2g = pd.read_csv('PageRank_500K_erdos_2g_WR.csv')
graphx_4g = pd.read_csv('PageRank_500K_erdos_4g_WR.csv')
flink_2g = pd.read_csv('PageRank_500K_erdos_2g_WR.csv')
flink_4g = pd.read_csv('PageRank_500K_erdos_Flink_4g_WR.csv')

plt.plot(graphx_2g["Time"], graphx_2g[" CPU %"], label = 'GraphX - 2GB')
plt.plot(graphx_4g["Time"], graphx_4g[" CPU %"], label = 'GraphX - 4GB')
plt.plot(flink_2g["Time"], flink_2g[" CPU %"], label = 'Flink - 2GB')
plt.plot(flink_4g["Time"], flink_4g[" CPU %"], label = 'Flink - 4GB')

plt.title("PageRank - 500K Erdos-Renyi")
plt.ylabel("CPU Usage (%)")
plt.xlabel("Time (sec)")

plt.legend()
plt.show()
```
